#2. Divisione del Dataset: Training, Validation e Test Set





La suddivisione del dataset è stata effettuata considerando due approcci alternativi:

* **Divisione per centro**

* **Divisione random**

La rete neurale è stata addestrata con gli stessi parametri in entrambi gli scenari. Dall’analisi dei risultati (presenti nella relazione)  è emerso che la **divisione random** consentiva una migliore generalizzazione e prestazioni superiori su nuove immagini rispetto alla divisione per centro. Per questo motivo si è scelto di adottare tale strategia nei successivi esperimenti.

```python
#codice ottimizzato per la divisione del dataset ( le modifiche: resize, strip suffix e conversione a png devono essere messe nel preprocessing della inferenza. Ora il dataset è gìa diviso con quelle modifiche quindi solo lasciamo il codice della divisione ma non lo runniamo di nuovo )

import os
import shutil
from pathlib import Path
import numpy as np
from sklearn.model_selection import train_test_split
from tqdm.auto import tqdm

# ========================= CONFIGURAZIONE =========================
PERC_TRAIN = 0.8  # 80% per il train
PERC_VAL = 0.2    # 20% per il validation

# Sequenze specifiche per ogni set
SEQ_TRAIN = {1, 5, 6, 9, 13, 21, 12}
SEQ_VAL = {3, 4, 19, 20}
SEQ_TEST = {2}  # Solo la sequenza 2 per il test

# Percorsi principali
ROOT_PATH = Path("/content/drive/MyDrive/Colab Notebooks/PolypGen-EIM-24-25")
OUTPUT_PATH = ROOT_PATH / "dataset_split"

# ====================== FUNZIONI PRINCIPALI ======================
def prepara_struttura_cartelle():
    """Crea la struttura delle cartelle per l'output"""
    for subset in ['train', 'val', 'test']:
        for content in ['images', 'masks']:
            (OUTPUT_PATH / subset / content).mkdir(parents=True, exist_ok=True)
    print("Struttura cartelle creata")

def processa_centri():
    """Elabora i centri C1-C6 dividendo le immagini in train/val"""
    print("\nElaborazione centri C1-C6...")

    # Raccogli tutte le immagini dai centri
    img_paths = []
    for c in range(1, 7):
        centro_path = ROOT_PATH / f"data_C{c}" / f"images_C{c}"
        if centro_path.exists():
            img_paths.extend(list(centro_path.glob("*.*")))

    # Divisione train/val
    train_paths, val_paths = train_test_split(
        img_paths, test_size=PERC_VAL, random_state=42
    )

    # Copia file per i centri
    for subset, paths in [('train', train_paths), ('val', val_paths)]:
        for img_path in tqdm(paths, desc=f"Copiando centri -> {subset}"):
            # Copia immagine
            dest_img = OUTPUT_PATH / subset / "images" / img_path.name
            shutil.copy2(img_path, dest_img)

            # Copia maschera corrispondente
            centro_id = img_path.parent.name.split("_")[-1]
            mask_path = img_path.parent.parent / f"masks_C{centro_id}" / img_path.name
            if mask_path.exists():
                dest_mask = OUTPUT_PATH / subset / "masks" / img_path.name
                shutil.copy2(mask_path, dest_mask)

    return len(train_paths), len(val_paths)

def processa_sequenze():
    """Elabora le sequenze assegnandole ai set specifici"""
    print("\nElaborazione sequenze...")
    conteggi = {'train': 0, 'val': 0, 'test': 0}

    for seq_id in SEQ_TRAIN | SEQ_VAL | SEQ_TEST:
        # Determina il subset di appartenenza
        if seq_id in SEQ_TRAIN:
            subset = 'train'
        elif seq_id in SEQ_VAL:
            subset = 'val'
        else:
            subset = 'test'

        seq_path = ROOT_PATH / "sequenceData" / f"seq{seq_id}"
        img_dir = seq_path / f"images_seq{seq_id}"
        mask_dir = seq_path / f"masks_seq{seq_id}"

        if not img_dir.exists() or not mask_dir.exists():
            continue

        print(f"  Copiando seq{seq_id} -> {subset}")
        img_count = 0

        # Copia tutte le immagini della sequenza
        for img_file in tqdm(list(img_dir.glob("*.*")), desc=f"Seq {seq_id}"):
            # Copia immagine
            dest_img = OUTPUT_PATH / subset / "images" / img_file.name
            shutil.copy2(img_file, dest_img)

            # Copia maschera corrispondente
            mask_file = mask_dir / img_file.name
            if mask_file.exists():
                dest_mask = OUTPUT_PATH / subset / "masks" / img_file.name
                shutil.copy2(mask_file, dest_mask)

            img_count += 1

        conteggi[subset] += img_count

    return conteggi['train'], conteggi['val'], conteggi['test']

def genera_report(centri_train, centri_val, seq_train, seq_val, seq_test):
    """Genera un report dettagliato della divisione"""
    total_train = centri_train + seq_train
    total_val = centri_val + seq_val
    total_test = seq_test

    print("\n" + "="*50)
    print("REPORT FINALE DELLA DIVISIONE DEL DATASET")
    print("="*50)

    print(f"\n{'SET':<10} | {'CENTRI':<10} | {'SEQUENZE':<10} | {'TOTALE':<10}")
    print("-"*45)
    print(f"{'TRAIN':<10} | {centri_train:<10} | {seq_train:<10} | {total_train:<10}")
    print(f"{'VAL':<10} | {centri_val:<10} | {seq_val:<10} | {total_val:<10}")
    print(f"{'TEST':<10} | {'-':<10} | {seq_test:<10} | {total_test:<10}")
    print("-"*45)
    print(f"{'TOTALE':<10} | {centri_train+centri_val:<10} | {seq_train+seq_val+seq_test:<10} | "
          f"{total_train+total_val+total_test:<10}")

    print("\nDETTAGLIO SEQUENZE:")
    print(f"  Train: {', '.join(str(s) for s in sorted(SEQ_TRAIN))}")
    print(f"  Val: {', '.join(str(s) for s in sorted(SEQ_VAL))}")
    print(f"  Test: {', '.join(str(s) for s in sorted(SEQ_TEST))}")

# ====================== ESECUZIONE PRINCIPALE ======================
if __name__ == "__main__":
    # Crea struttura cartelle
    prepara_struttura_cartelle()

    # Elabora centri (C1-C6)
    centri_train, centri_val = processa_centri()

    # Elabora sequenze
    seq_train, seq_val, seq_test = processa_sequenze()

    # Genera report finale
    genera_report(centri_train, centri_val, seq_train, seq_val, seq_test)

    print("\nOperazione completata con successo!")
    print(f"Dataset organizzato in: {OUTPUT_PATH}")
```

```python
# --- Funzione di conversione in-place JPG → PNG ---
def jpg_to_png_inplace(img_path: Path, is_mask: bool):
    """
    Converte un .jpg in .png nello stesso folder.
    • is_mask=True  ➜ binarizza e salva valori 0/1
    • is_mask=False ➜ lascia a colori (RGB)
    Elimina il file .jpg originale.
    """
    if not img_path.suffix.lower() == ".jpg":
        return  # salta eventuali altri formati

    # nome file con estensione .png
    png_path = img_path.with_suffix(".png")

    # carica immagine
    img = Image.open(img_path)

    if is_mask:
        img = img.convert("L")
        arr = np.array(img)
        bin_arr = (arr > 127).astype(np.uint8)  # 0 / 1
        img = Image.fromarray(bin_arr, mode="L")
    else:
        img = img.convert("RGB")

    # salva .png, poi cancella il .jpg
    img.save(png_path, compress_level=3)
    img_path.unlink()
    print(f"Convertito: {img_path.name} → {png_path.name}")

# --- Elaborazione centri (data_C1 … data_C6) ---
for c in range(1, 7):
    imgs_dir  = ROOT / f"data_C{c}" / f"images_C{c}"
    masks_dir = ROOT / f"data_C{c}" / f"masks_C{c}"

    for p in tqdm(list(imgs_dir.glob("*.jpg")),  desc=f"C{c} imgs   JPG→PNG"):
        jpg_to_png_inplace(p, is_mask=False)
    for p in tqdm(list(masks_dir.glob("*.jpg")), desc=f"C{c} masks  JPG→PNG"):
        jpg_to_png_inplace(p, is_mask=True)

# --- Elaborazione sequenze (seq1 … seq23) ---
for s in range(1, 24):
    imgs_dir  = ROOT / "sequenceData" / f"seq{s}" / f"images_seq{s}"
    masks_dir = ROOT / "sequenceData" / f"seq{s}" / f"masks_seq{s}"

    for p in tqdm(list(imgs_dir.glob("*.jpg")),  desc=f"seq{s} imgs  JPG→PNG"):
        jpg_to_png_inplace(p, is_mask=False)
    for p in tqdm(list(masks_dir.glob("*.jpg")), desc=f"seq{s} masks JPG→PNG"):
        jpg_to_png_inplace(p, is_mask=True)

print("✔️  Conversione JPG → PNG completata con maschere binarie 0/1.")
```

```python
## =========================== DEBUG =========================================
import os
base_path = "/content/drive/MyDrive/Colab Notebooks/PolypGen-EIM-24-25"

def summarize_folder(image_folder, mask_folder, name=""):
    image_files = sorted([
        f for f in os.listdir(image_folder) if f.lower().endswith((".jpg", ".png"))
    ])
    mask_files = sorted([
        f for f in os.listdir(mask_folder) if f.lower().endswith((".jpg", ".png"))
    ])

    n_images = len(image_files)
    n_masks = len(mask_files)

    if n_images == 0 or n_masks == 0:
        print(f"❌ {name}: no images or masks found")
        return

    sizes = []
    dtypes = set()
    mask_mins = []
    mask_maxs = []

    # Check for name correspondence
    img_names_base = set(os.path.splitext(f)[0] for f in image_files)
    mask_names_base = set(os.path.splitext(f)[0] for f in mask_files)

    matched = img_names_base & mask_names_base
    only_images = img_names_base - mask_names_base
    only_masks = mask_names_base - img_names_base

    # Use only the paired files
    paired = sorted(list(matched))
    for base_name in paired:
        img_path = os.path.join(image_folder, base_name + ".png")
        if not os.path.exists(img_path):
            img_path = os.path.join(image_folder, base_name + ".jpg")
        mask_path = os.path.join(mask_folder, base_name + ".png")
        if not os.path.exists(mask_path):
            mask_path = os.path.join(mask_folder, base_name + ".jpg")

        try:
            img = Image.open(img_path)
            sizes.append(img.size)

            mask = Image.open(mask_path)
            mask_np = np.asarray(mask)
            dtypes.add(str(mask_np.dtype))
            mask_mins.append(mask_np.min())
            mask_maxs.append(mask_np.max())
        except Exception as e:
            print(f"⚠️ Error with file {base_name}: {e}")
            continue

    if sizes:
        avg_width = int(np.mean([s[0] for s in sizes]))
        avg_height = int(np.mean([s[1] for s in sizes]))
        min_val = int(np.min(mask_mins))
        max_val = int(np.max(mask_maxs))
    else:
        avg_width = avg_height = min_val = max_val = 0

    print(f"\n📁 {name}")
    print(f"    Total images: {n_images}, Total masks: {n_masks}")
    print(f"    Successfully paired: {len(paired)}")
    if only_images:
        print(f"    ⚠️ {len(only_images)} images without a corresponding mask")
    if only_masks:
        print(f"    ⚠️ {len(only_masks)} masks without a corresponding image")
    print(f"    Average size: {avg_width}x{avg_height}, dtype(s): {dtypes}, mask min: {min_val}, max: {max_val}")

# Centers C1–C6
for c in range(1, 7):
    images_path = os.path.join(base_path, f"data_C{c}", f"images_C{c}")
    masks_path = os.path.join(base_path, f"data_C{c}", f"masks_C{c}")
    summarize_folder(images_path, masks_path, name=f"C{c}")

# Sequences 1–23
for s in range(1, 24):
    images_path = os.path.join(base_path, "sequenceData", f"seq{s}", f"images_seq{s}")
    masks_path = os.path.join(base_path, "sequenceData", f"seq{s}", f"masks_seq{s}")
    summarize_folder(images_path, masks_path, name=f"Seq{s}")
```

# Divisione Random del Dataset

```python
# Configurazione e Costanti
ROOT_PATH = Path("/content/drive/MyDrive/Colab Notebooks/PolypGen-EIM-24-25")

# IMPORTANTE! Nuovo percorso di output per il dataset suddiviso
OUTPUT_DATASET_PATH = ROOT_PATH / "dataset_split"

#Proporzioni per la divisione del dataset (Addestramento e Validazione)
# Assicurati che TRAIN_SIZE + VAL_SIZE = 1.0
TRAIN_SIZE = 0.80 # 80% per l’addestramento
VAL_SIZE = 0.20   # 20% per la validazione

def create_dataset_folders(base_path: Path):
  """Crea la struttura delle cartelle train/val con sottocartelle images/masks."""
  for subset in ["train", "val"]:
        (base_path / subset / "images").mkdir(parents=True, exist_ok=True)
        (base_path / subset / "masks").mkdir(parents=True, exist_ok=True)
        print(f"Struttura della cartella creata in: {base_path}")

def copy_image_and_mask(original_img_path: Path, destination_base_path: Path) -> tuple[bool, bool, str]:
    """
    Copia un'immagine e la sua maschera corrispondente nella cartella di destinazione.
    Restituisce (immagine_copiata_ok, maschera_copiata_ok, messaggio).
    """
    mask_name = original_img_path.name
    center_id_part = original_img_path.parent.name.replace("images_", "")
    original_mask_path = original_img_path.parent.parent / f"masks_{center_id_part}" / mask_name

    img_copied = False
    mask_copied = False
    message = ""

    dest_img_path = destination_base_path / "images" / original_img_path.name
    dest_mask_path = destination_base_path / "masks" / original_mask_path.name

    if not original_img_path.exists():
        message = f"ERRORE: Immagine originale non trovata: {original_img_path}"
        return False, False, message

    try:
        shutil.copy2(original_img_path, dest_img_path)
        img_copied = True
    except Exception as e:
        message = f"ERRORE: Impossibile copiare l'immagine {original_img_path}: {e}"
        return False, False, message

    if original_mask_path.exists():
        try:
            shutil.copy2(original_mask_path, dest_mask_path)
            mask_copied = True
        except Exception as e:
            message = f"ERRORE: Impossibile copiare la maschera {original_mask_path}: {e}"
            return img_copied, False, message
    else:
        # Questo è previsto per le immagini 'NO' (senza polipo)
        message = f"ATTENZIONE: Maschera non trovata per {original_img_path} (atteso se è un'immagine senza polipo)."

    return img_copied, mask_copied, message


def create_random_split_dataset(root_data_path: Path, output_base_path: Path,
                                train_ratio: float, val_ratio: float):

    all_image_paths = []
    print("Raccogliendo tutti i percorsi di immagini dei centri...")
    center_dirs = [p for p in root_data_path.glob("data_C*") if p.is_dir()]
    for center_dir in center_dirs:
        center_id_short = center_dir.name.replace("data_", "")
        images_dir = center_dir / f"images_{center_id_short}"
        if images_dir.is_dir():
            all_image_paths.extend(list(images_dir.glob("*.png")))
        else:
            print(f"ATTENZIONE: Directory di immagini non trovata per {center_dir.name}: {images_dir}")

    if not all_image_paths:
        print("Non sono state trovate immagini in nessuna directory del centro. Uscendo.")
        return None, None

    print(f"Sono state trovate {len(all_image_paths)} immagini in totale.")

    df_all_images = pd.DataFrame({'path': [str(p) for p in all_image_paths]})

    df_train_paths, df_val_paths = train_test_split(
        df_all_images,
        test_size=val_ratio / (train_ratio + val_ratio),
        random_state=42
    )

    print(f"\nDivisione del Dataset:")
    print(f"  Set di addestramento: {len(df_train_paths)} immagini ({train_ratio*100:.0f}%)")
    print(f"  Set di validazione: {len(df_val_paths)} immagini ({val_ratio*100:.0f}%)")

    create_dataset_folders(output_base_path)

    # --- Contatori per il report ---
    train_images_copied = 0
    train_masks_copied = 0
    train_copy_errors = []
    train_mask_missing_warnings = []

    val_images_copied = 0
    val_masks_copied = 0
    val_copy_errors = []
    val_mask_missing_warnings = []


    print("\nCopia di immagini e maschere per il set di addestramento...")
    for index, row in tqdm(df_train_paths.iterrows(), total=len(df_train_paths), desc="Copiando Train"):
        img_copied, mask_copied, message = copy_image_and_mask(Path(row['path']), output_base_path / "train")
        if img_copied:
            train_images_copied += 1
        if mask_copied:
            train_masks_copied += 1
        if "ERRORE" in message:
            train_copy_errors.append(message)
        elif "ATTENZIONE" in message:
            train_mask_missing_warnings.append(message)

    print("\nCopia di immagini e maschere per il set di validazione...")
    for index, row in tqdm(df_val_paths.iterrows(), total=len(df_val_paths), desc="Copiando Val"):
        img_copied, mask_copied, message = copy_image_and_mask(Path(row['path']), output_base_path / "val")
        if img_copied:
            val_images_copied += 1
        if mask_copied:
            val_masks_copied += 1
        if "ERRORE" in message:
            val_copy_errors.append(message)
        elif "ATTENZIONE" in message:
            val_mask_missing_warnings.append(message)

    print("\n--- Rapporto sul processo di copia ---")
    print(f"**Set di Addestramento ('train'):**")
    print(f"  Immagini elaborate attese: {len(df_train_paths)}")
    print(f"  Immagini copiate correttamente: {train_images_copied}")
    print(f"  Maschere copiate correttamente: {train_masks_copied}")
    if train_copy_errors:
        print(f"  ERRORI di copia di immagini/maschere in 'train': {len(train_copy_errors)}")
        for error in train_copy_errors:
            print(f"    - {error}")
    if train_mask_missing_warnings:
        print(f"  AVVISI sulle maschere non trovate in 'train': {len(train_mask_missing_warnings)}")
        # Puoi scegliere di stampare solo un riepilogo o i primi N per non saturare
        # for warning in train_mask_missing_warnings[:5]: # Mostrare solo le prime 5
        #     print(f"    - {warning}")

    print(f"\n**Set di Validazione ('val'):**")
    print(f"  Immagini elaborate attese: {len(df_val_paths)}")
    print(f"  Immagini copiate correttamente: {val_images_copied}")
    print(f"  Maschere copiate correttamente: {val_masks_copied}")
    if val_copy_errors:
        print(f"  ERRORI di copia immagini/maschere in 'val': {len(val_copy_errors)}")
        for error in val_copy_errors:
            print(f"    - {error}")
    if val_mask_missing_warnings:
        print(f"   AVVISI sulle maschere non trovate in'val': {len(val_mask_missing_warnings)}")
        # for warning in val_mask_missing_warnings[:5]: # Mostrar solo las primeras 5
        #     print(f"    - {warning}")

    total_images_expected = len(df_train_paths) + len(df_val_paths)
    total_images_copied = train_images_copied + val_images_copied
    total_masks_copied = train_masks_copied + val_masks_copied
    total_errors = len(train_copy_errors) + len(val_copy_errors)

    print(f"\n**RRiepilogo Generale:**")
    print(f"  Totale immagini previste da copiare: {total_images_expected}")
    print(f"  Totale immagini copiate correttamente: {total_images_copied}")
    print(f"  Totale maschere copiate correttamente: {total_masks_copied}")
    print(f"  Totale errori di copia: {total_errors}")

    if total_images_copied == total_images_expected and total_errors == 0:
        print("\nTutte le immagini sono state copiate correttamente con le relative maschere (se presenti)!")
    elif total_errors > 0:
        print("\nATTENZIONE! Si sono verificati errori durante la copia di immagini o maschere. Controlla i messaggi di errore.")
    else:
        print("\nProcesso di copia completato con avvisi o alcune immagini/maschere non copiate. Controlla il report.")

    return df_train_paths, df_val_paths

# --- Eseguire la creazione del dataset ---
if __name__ == "__main__":
    df_train_created, df_val_created = create_random_split_dataset(
        root_data_path=ROOT_PATH,
        output_base_path=OUTPUT_DATASET_PATH,
        train_ratio=TRAIN_SIZE,
        val_ratio=VAL_SIZE
    )
```

```python
from tqdm import tqdm

# --- Configurazione e Costanti ---
ROOT_PATH = Path("/content/drive/MyDrive/Colab Notebooks/PolypGen-EIM-24-25")

# Percorso in cui si trovano le cartelle 'seq<i_number>' con images_seq<i_number> e masks_seq<i_number>
SEQUENCE_DATA_ROOT = ROOT_PATH / "sequenceData"

# Percorso di output per il dataset suddiviso (dove sono state create le cartelle train/val/images/masks)
OUTPUT_DATASET_PATH = ROOT_PATH / "dataset_split"

# --- Definizione delle Sequenze per Train e Val ---
# Questi sono gli ID delle sequenze (i numeri 'i' in seq<i>) che vuoi in ciascun insieme.
TRAIN_SEQUENCES = {1, 5, 6, 9, 13, 21, 12}
VAL_SEQUENCES = {3, 4, 19, 20}

# Verifica sovrapposizioni per evitare errori logici
if not TRAIN_SEQUENCES.isdisjoint(VAL_SEQUENCES):
    print("ATTENZIONE: Le sequenze di formazione e di convalida si sovrappongono! Si prega di controllare le vostre liste.")
      # Considera di usare exit() qui se è un errore critico per il tuo pipeline.

# --- Funzioni Ausiliarie ---

def create_dataset_folders(base_path: Path):
    """Crea la struttura delle cartelle train/val con sottocartelle images/masks."""
    for subset in ["train", "val"]:
        (base_path / subset / "images").mkdir(parents=True, exist_ok=True)
        (base_path / subset / "masks").mkdir(parents=True, exist_ok=True)
    print(f"Struttura della cartella creata in: {base_path}")

def copy_contents_of_folders(source_images_folder: Path, source_masks_folder: Path,
                             dest_images_folder: Path, dest_masks_folder: Path) -> tuple[int, int, list, list]:
    """
    Copia tutte le immagini e le relative maschere dalle cartelle di origine a quelle di destinazione.
    Restituisce il numero di immagini/maschere copiate e le liste di errori/avvisi.
    """
    images_copied = 0
    masks_copied = 0
    copy_errors = []
    mask_missing_warnings = []

    if not source_images_folder.is_dir():
        copy_errors.append(f"ERRORE: Cartella delle immagini di origine non trovata: {source_images_folder}")
        return images_copied, masks_copied, copy_errors, mask_missing_warnings

    print(f"Copiando immagini da {source_images_folder.name}...")
    image_files = list(source_images_folder.glob("*.png")) + list(source_images_folder.glob("*.jpg")) # Considera sia png che jpg
    for img_path in tqdm(image_files, desc=f"Copiando {source_images_folder.name}"):
        dest_img_path = dest_images_folder / img_path.name
        mask_path = source_masks_folder / img_path.name # La maschera deve avere lo stesso nome dell'immagine
        dest_mask_path = dest_masks_folder / img_path.name

        try:
            shutil.copy2(img_path, dest_img_path)
            images_copied += 1
        except Exception as e:
            copy_errors.append(f"ERRORE durante la copia dell'immagine {img_path.name}: {e}")
            continue # Si la imagen falla, no intentes copiar la máscara

        if mask_path.exists():
            try:
                shutil.copy2(mask_path, dest_mask_path)
                masks_copied += 1
            except Exception as e:
                copy_errors.append(f"ERRORE durante la copia della masxhera {mask_path.name}: {e}")
        else:
            mask_missing_warnings.append(f"ATTENZIONE: Maschera {mask_path.name} non trovata per l'immagine {img_path.name} (normale se è un'immagine senza polipo).")

    return images_copied, masks_copied, copy_errors, mask_missing_warnings


def split_dataset_by_sequence_folders(sequence_data_root: Path, output_base_path: Path,
                                       train_sequences: set, val_sequences: set):

    print("Preparazione della divisione del dataset per le cartelle di sequenza...")

    # Assicurarsi che le cartelle di destinazione esistano
    create_dataset_folders(output_base_path)

    # --- Contatori e liste per il report ---
    total_train_images_copied = 0
    total_train_masks_copied = 0
    all_train_copy_errors = []
    all_train_mask_missing_warnings = []

    total_val_images_copied = 0
    total_val_masks_copied = 0
    all_val_copy_errors = []
    all_val_mask_missing_warnings = []

    processed_sequences_count = 0
    unassigned_sequences_count = 0

    # Itera attraverso le cartelle 'seq<i>'
    for seq_dir in sorted(list(sequence_data_root.glob("seq*"))):
        if not seq_dir.is_dir():
            continue

        # Estrae l'ID della sequenza dal nome della cartella (es. 'seq1' -> 1)
        try:
            seq_id = int(seq_dir.name.replace("seq", ""))
        except ValueError:
            print(f"AVVISO: La cartella '{seq_dir.name}' non ha un formato di sequenza valido (seq<number>). Ignorata.")
            continue

        source_images_folder = seq_dir / f"images_seq{seq_id}"
        source_masks_folder = seq_dir / f"masks_seq{seq_id}"

        if not source_images_folder.is_dir() or not source_masks_folder.is_dir():
            print(f"AVVISO: Cartelle images_seq{seq_id} o masks_seq{seq_id} non trovate in {seq_dir.name}. Sequenza ignorata.")
            unassigned_sequences_count += 1
            continue

        target_subset = None
        if seq_id in train_sequences:
            target_subset = "train"
        elif seq_id in val_sequences:
            target_subset = "val"
        else:
            unassigned_sequences_count += 1
            print(f"La sequenza '{seq_id}' non è assegnata né a train né a val. Verrà ignorata.")
            continue # Salta se la sequenza non è in nessuna lista

        processed_sequences_count += 1
        print(f"\nElaborazione della sequenza {seq_id} per il set {target_subset}...")

        dest_images_folder = output_base_path / target_subset / "images"
        dest_masks_folder = output_base_path / target_subset / "masks"

        images_copied, masks_copied, errors, warnings = copy_contents_of_folders(
            source_images_folder, source_masks_folder,
            dest_images_folder, dest_masks_folder
        )

        if target_subset == "train":
            total_train_images_copied += images_copied
            total_train_masks_copied += masks_copied
            all_train_copy_errors.extend(errors)
            all_train_mask_missing_warnings.extend(warnings)
        elif target_subset == "val":
            total_val_images_copied += images_copied
            total_val_masks_copied += masks_copied
            all_val_copy_errors.extend(errors)
            all_val_mask_missing_warnings.extend(warnings)

    print("\n--- Report Finale del Processo di Copia ---")
    print(f"**Riepilogo delle Sequenze:**")
    print(f"  Sequenze elaborate (assegnate a train/val): {processed_sequences_count}")
    print(f"  Sequenze non assegnate o incomplete (ignorate): {unassigned_sequences_count}")

    print(f"\n**Set di Addestramento ('train'):**")
    print(f"  Totale immagini copiate: {total_train_images_copied}")
    print(f"  Totale maschere copiate: {total_train_masks_copied}")
    if all_train_copy_errors:
        print(f"  ERRORI di copia in 'train': {len(all_train_copy_errors)}")
        for error in all_train_copy_errors:
            print(f"    - {error}")
    if all_train_mask_missing_warnings:
        print(f"  AVVISI di maschere non trovate in 'train': {len(all_train_mask_missing_warnings)}")
        # for warning in all_train_mask_missing_warnings: print(f"    - {warning}") # Rimuovi commento per vedere tutti gli avvisi


    print(f"\n**Set di Validazione ('val'):**")
    print(f"  Totale immagini copiate: {total_val_images_copied}")
    print(f"  Totale maschere copiate: {total_val_masks_copied}")
    if all_val_copy_errors:
        print(f"  ERRORI di copia in 'val': {len(all_val_copy_errors)}")
        for error in all_val_copy_errors:
            print(f"    - {error}")
    if all_val_mask_missing_warnings:
        print(f"  AVVISI di maschere non trovate in 'val': {len(all_val_mask_missing_warnings)}")
        # for warning in all_val_mask_missing_warnings: print(f"    - {warning}") #Rimuovi commento per vedere tutti gli avvisi

    total_errors_overall = len(all_train_copy_errors) + len(all_val_copy_errors)
    if total_errors_overall == 0 and processed_sequences_count > 0:
        print("\nCopia dei dati per sequenza completata con successo!")
    elif total_errors_overall > 0:
        print("\nATTENZIONE! Si sono verificati errori durante la copia. Controlla i messaggi di errore dettagliati.")
    else:
        print("\nProcesso completato, ma nessuna sequenza è stata elaborata o sono stati trovati avvisi. Controlla il report.")


# --- Eseguire la creazione del dataset ---
if __name__ == "__main__":
    # Montare Google Drive se non è ancora montato (necessario solo in Colab)
    # from google.colab import drive
    # drive.mount('/content/drive')
    # print("Google Drive montato.")

    split_dataset_by_sequence_folders(
        sequence_data_root=SEQUENCE_DATA_ROOT,
        output_base_path=OUTPUT_DATASET_PATH,
        train_sequences=TRAIN_SEQUENCES,
        val_sequences=VAL_SEQUENCES
    )
```

```python

# L’unica sequenza che vogliamo usare per il test
TEST_SEQUENCE_ID = 2

# --- Funzioni ausiliarie ---

def create_test_folders(base_path: Path):
    """Crea la struttura di cartelle per il test con le sottocartelle images/masks."""
    (base_path / "test" / "images").mkdir(parents=True, exist_ok=True)
    (base_path / "test" / "masks").mkdir(parents=True, exist_ok=True)
    print(f"Struttura della cartella creata in: {base_path}")

def copy_contents_of_folders(source_images_folder: Path, source_masks_folder: Path,
                             dest_images_folder: Path, dest_masks_folder: Path) -> tuple[int, int, list, list]:
    """
    Copia tutte le immagini e le loro maschere dalle cartelle di origine a quelle di destinazione.
    Restituisce il numero di immagini/maschere copiate e le liste di errori/avvisi.
    """
    images_copied = 0
    masks_copied = 0
    copy_errors = []
    mask_missing_warnings = []

    if not source_images_folder.is_dir():
        copy_errors.append(f"ERRORE: Directory del immagini di origine non travata: {source_images_folder}")
        return images_copied, masks_copied, copy_errors, mask_missing_warnings

    print(f"Copiando immagini da {source_images_folder.name}...")
    image_files = list(source_images_folder.glob("*.png")) + list(source_images_folder.glob("*.jpg")) #  Considera i file png e jpg
    for img_path in tqdm(image_files, desc=f"Copiando {source_images_folder.name}"):
        dest_img_path = dest_images_folder / img_path.name
        mask_path = source_masks_folder / img_path.name # La maschera deve avere lo stesso nome dell'immagine
        dest_mask_path = dest_masks_folder / img_path.name

        try:
            shutil.copy2(img_path, dest_img_path)
            images_copied += 1
        except Exception as e:
            copy_errors.append(f"ERRORE nella copia dell'immagine {img_path.name}: {e}")
            continue # Se l'immagine fallisce, non provare a copiare la maschera

        if mask_path.exists():
            try:
                shutil.copy2(mask_path, dest_mask_path)
                masks_copied += 1
            except Exception as e:
                copy_errors.append(f"ERRORE nella copia della maschera {mask_path.name}: {e}")
        else:
            mask_missing_warnings.append(f"AVVISO: Maschera {mask_path.name} non trovata per l'immagine {img_path.name} (normale se è un'immagine senza polipo).")

    return images_copied, masks_copied, copy_errors, mask_missing_warnings


def move_sequence_to_test(sequence_data_root: Path, output_base_path: Path, test_sequence_id: int):

    print(f"Preparazione allo spostamento della sequenza {test_sequence_id} nel set di test...")

    # Assicurarsi che le cartelle di destinazione esistano
    create_test_folders(output_base_path)

    # Percorsi specifici per la sequenza di test
    seq_dir = sequence_data_root / f"seq{test_sequence_id}"
    source_images_folder = seq_dir / f"images_seq{test_sequence_id}"
    source_masks_folder = seq_dir / f"masks_seq{test_sequence_id}"

    dest_images_folder = output_base_path / "test" / "images"
    dest_masks_folder = output_base_path / "test" / "masks"

    # --- Contatori e elenchi per il report ---
    total_test_images_copied = 0
    total_test_masks_copied = 0
    all_test_copy_errors = []
    all_test_mask_missing_warnings = []

    if not seq_dir.is_dir():
        print(f "ERRORE: La cartella della sequenza '{seq_dir}' non esiste. Verifica il percorso e il nome della cartella.")
        return

    if not source_images_folder.is_dir() or not source_masks_folder.is_dir():
        print(f"ERRORE: Le cartelle 'images_seq{test_sequence_id}' o 'masks_seq{test_sequence_id}' non sono state trovate all’interno di '{seq_dir}'.")
        print("Assicurati che la struttura sia del tipo: .../sequenceData/seq2/images_seq2 e .../sequenceData/seq2/masks_seq2.")
        return

    print(f"\nElaborazione della sequenza {test_sequence_id} per il set di test...")

    images_copied, masks_copied, errors, warnings = copy_contents_of_folders(
        source_images_folder, source_masks_folder,
        dest_images_folder, dest_masks_folder
    )

    total_test_images_copied += images_copied
    total_test_masks_copied += masks_copied
    all_test_copy_errors.extend(errors)
    all_test_mask_missing_warnings.extend(warnings)

    print("\n--- Report finale del processo di copia ---")
    print(f"**Set di Test ('test'):**")
    print(f"  Sequenza processata: {test_sequence_id}")
    print(f"  Totale immagini copiate: {total_test_images_copied}")
    print(f"  Totale maschere copiate: {total_test_masks_copied}")
    if all_test_copy_errors:
        print(f"  ERRORI di copia nel 'test': {len(all_test_copy_errors)}")
        for error in all_test_copy_errors:
            print(f"    - {error}")
    if all_test_mask_missing_warnings:
        print(f"  AVVISI: Maschere non trovate nel 'test': {len(all_test_mask_missing_warnings)}")
        # for warning in all_test_mask_missing_warnings: print(f"    - {warning}") # Decommentare per visualizzare tutti gli avvisi

    total_errors_overall = len(all_test_copy_errors)
    if total_errors_overall == 0 and total_test_images_copied > 0:
        print(f"\nSequenza {test_sequence_id} copiata con successo nel set di test!")
    elif total_errors_overall > 0:
        print("\nATTENZIONE! Sono stati rilevati errori durante la copia. Controlla i messaggi di errore dettagliati.")
    else:
        print(f"\nNessuna immagine è stata copiata per la sequenza {test_sequence_id}. Controlla i messaggi di errore o avviso.")


# --- Eseguire la copia della sequenza di test ---
if __name__ == "__main__":
    # Montare Google Drive se non è già montato (necessario solo in Colab)
    # from google.colab import drive
    # drive.mount('/content/drive')
    # print("Google Drive montado.")

    move_sequence_to_test(
        sequence_data_root=SEQUENCE_DATA_ROOT,
        output_base_path=OUTPUT_DATASET_PATH,
        test_sequence_id=TEST_SEQUENCE_ID
    )
```

```python

# --- Configurazione ---
DATASET_ROOT = Path("/content/drive/MyDrive/Colab Notebooks/PolypGen-EIM-24-25/dataset_split")

# ---Funzione per analizzare un sottoinsieme (train, val, test) ---
def analyze_subset(subset_path: Path):
    """
    Analizza le immagini e le maschere in un dato sottoinsieme (train, val, test)
    e restituisce il conteggio delle immagini totali, positive e negative.
    """
    images_dir = subset_path / "images"
    masks_dir = subset_path / "masks"

    if not images_dir.is_dir():
        print(f"Avviso: Directory delle immagini non trovata per {subset_path.name}. Saltando.")
        return 0, 0, 0

    all_images = sorted(list(images_dir.glob("*.png")) + list(images_dir.glob("*.jpg")))
    total_images = len(all_images)
    positive_images = 0
    negative_images = 0

    print(f"\nAnalizzando il sottoinsieme: {subset_path.name}...")
    for img_path in tqdm(all_images, desc=f"Elaborando {subset_path.name}"):
        mask_path = masks_dir / img_path.name # Si assume che la maschera abbia lo stesso nome dell’immagine

        if mask_path.exists():
            try:
                mask = Image.open(mask_path).convert('L') # Apre la maschera in scala di grigi
                mask_np = np.array(mask)

                # Una maschera è "positiva" se contiene almeno un pixel non nullo (indica presenza di polipo)
                if np.any(mask_np > 0):
                    positive_images += 1
                else:
                    negative_images += 1
            except Exception as e:
                print(f"Errore durante l’elaborazione della maschera {mask_path.name}: {e}. Verrà considerata come 'negativa' per precauzione.")
                negative_images += 1 # In caso di errore, si assume negativa (o si gestisce diversamente)
        else:
           # Se non esiste maschera, si considera l’immagine come "negativa" (senza polipo)
           # È quanto concordato precedentemente per immagini senza polipi
            negative_images += 1

    return total_images, positive_images, negative_images

# --- Esecuzione dell’analisi ---
if __name__ == "__main__":
    # Monta Google Drive se non è montato (necessario solo in Colab)
    # from google.colab import drive
    # drive.mount('/content/drive')
    # print("Google Drive mounted.")

    print(f"Inizio dell’analisi del dataset in: {DATASET_ROOT}")

    subsets = ["train", "val", "test"]
    results = {}

    for subset in subsets:
        subset_path = DATASET_ROOT / subset
        if subset_path.is_dir():
            total, positive, negative = analyze_subset(subset_path)
            results[subset] = {
                "total_images": total,
                "positive_images": positive,
                "negative_images": negative
            }
        else:
            print(f"Avviso: La cartella '{subset}' non è stata trovata in {DATASET_ROOT}. Saltando.")

    print("\n--- Riepilogo dell’organizzazione del Dataset ---")
    for subset, data in results.items():
        print(f"\nSottoinsieme: {subset.upper()}")
        print(f"  Immagini totali: {data['total_images']}")
        print(f"  Immagini con polipo (positive): {data['positive_images']}")
        print(f"  Immagini senza polipo (negative): {data['negative_images']}")
        if data['total_images'] > 0:
            pos_percentage = (data['positive_images'] / data['total_images']) * 100
            neg_percentage = (data['negative_images'] / data['total_images']) * 100
            print(f"  Percentuale positive: {pos_percentage:.2f}%")
            print(f"  Percentuale negative: {neg_percentage:.2f}%")
        else:
            print("  Nessuna immagine in questo sottoinsieme.")

    print("\nAnalisi completata.")
```

```python
''' RIMUOVE SUFFISSO _MASK '''

def strip_mask_suffix(cartella_mask: str):
    """Rimuove il suffisso '_mask' dai file nella cartella specificata"""
    cartella = Path(cartella_mask)
    if not cartella.is_dir():
        return
    for percorso_file in cartella.glob("*_mask.*"):
        nuovo_nome = percorso_file.name.replace("_mask", "")
        nuovo_path = percorso_file.with_name(nuovo_nome)
        if nuovo_path.exists():
            print(f"Avviso: {nuovo_path.name} esiste già. Salto...")
            continue
        percorso_file.rename(nuovo_path)
        print(f"Rinominato: {percorso_file.name} → {nuovo_path.name}")


''' LOAD IMAGE & MASK: cv2 imread '''

def load_img_mask(image_path, mask_path):
    # image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # RGB
    # maschera
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    #mask = (mask > 128).astype(np.uint8) # binarized 0/1
    return image, mask

''' RESIZE: resize skimage transform, binarizzazione maschera '''

def resize_im(image_array, target_size, order, anti_aliasing):
    # Ridimensionamento della maschera (interpolazione nearest con order=0 e filtro anti_aliasing false),
    # Ridimensionamento dell'immagine (interpolazione bilineare con order=1 e filtro anti_aliasing)
    resized_img = skimage.transform.resize(image_array, target_size, anti_aliasing=anti_aliasing, order=order) # float64
    if order==0:
        resized_img = resized_im.astype(np.uint8) # binarized 0/1, uint8
    if order==1:
        resized_img = (resized_img * 255).astype(np.uint8)
    return resized_img

''' OMOGENEIZZAZIONE: REINHARD NORMALIZATION '''

def compute_global_stats(image_paths):
    lab_means = []
    lab_stds = []
    for path in tqdm(image_paths, desc="Calculating global statistics (Reinhard)"):
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        lab = color.rgb2lab(img)
        lab_means.append(np.mean(lab, axis=(0,1)))
        lab_stds.append(np.std(lab, axis=(0,1)))
    global_mean = np.mean(lab_means, axis=0)
    global_std = np.maximum(np.mean(lab_stds, axis=0), 1e-6)
    return global_mean, global_std

def compute_lab_stats(image_rgb):
    image_lab = color.rgb2lab(image_rgb)
    mean = image_lab.mean(axis=(0, 1))
    std = image_lab.std(axis=(0, 1))
    return mean, std

def apply_reinhard_normalization(image_rgb, target_mean, target_std, epsilon=1e-6):
    source_mean, source_std = compute_lab_stats(image_rgb)
    image_lab = color.rgb2lab(image_rgb)
    epsilon_adjusted = 1e-6
    safe_source_std = np.where(source_std < epsilon_adjusted, epsilon_adjusted, source_std)
    safe_target_std = np.where(target_std < epsilon_adjusted, epsilon_adjusted, target_std) # Anche il target_std non dovrebbe essere troppo piccolo
    norm_lab = (image_lab - source_mean) / safe_source_std * safe_target_std + target_mean # Usa safe_target_std qui
    norm_lab[..., 0] = np.clip(norm_lab[..., 0], 0, 100) # L channel
    norm_lab[..., 1] = np.clip(norm_lab[..., 1], -128, 127) # a channel
    norm_lab[..., 2] = np.clip(norm_lab[..., 2], -128, 127) # b channel
    img_rgb_normalized = color.lab2rgb(norm_lab)
    final_img_clipped = np.clip(img_rgb_normalized, 0, 1) # Clipping finale a [0,1] per RGB
    return final_img_clipped

''' ELIMINAZIONE DI ARTEFATTI LUMINOSI''' # dopo lo pulisco, da runnare una volta solo ver vedere i risultati ma non viene applicato nella pipeline finale

def eliminazione_artefatti_luminosi(img):
  # Step 1: Identificazione pixel con riflessi
    mask = step_1_identificazione_riflessi(img)
  # Step 2: Operazioni morfologiche
    grown_mask = step_2_highlights_mask_op_morfologici(img, mask, apertura=3, chiusura=6, dilatazione=3)
  # Step 3: Inpainting per rimozione riflessi
    img_no_highlights = step_3_highlights_inpainting(img, grown_mask, mask, True)
    return img_no_highlights

def step_1_identificazione_riflessi(img): # umbralizzazione.
    ''' Restituisce una maschera binaria dove 1 indica la presenza di riflessi '''
    image_hsv_uint8 = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2HSV) # Converte l'immagine in HSV (valori uint8)
    H, S, V = cv2.split(image_hsv_uint8)
    # parametri per l'identificazione dei riflessi
    V_threshold_low = 100 # Soglia di Valore/Intensità: i pixel con V > 230 sono molto luminosi.
    S_threshold_high = 40  # Soglia di Saturazione: i pixel con S < 30 hanno poca saturazione (quasi bianchi/grigi).
    # Crea le maschere per alte luminosità e basse saturazioni e combina le condizioni (AND)
    high_brightness_mask = V > V_threshold_low
    low_saturation_mask = S < S_threshold_high
    specular_mask = np.logical_and(high_brightness_mask, low_saturation_mask).astype(np.uint8) * 255 # Convertir a 0 o 255

def step_2_highlights_mask_op_morfologici(img, mask, apertura: int = 3, chiusura: int=6, dilatazione: int=3):
    ''' Applicazione di operazioni morfologiche per pulire la maschera dei riflessi '''
    closed_mask = morphology.binary_closing(morphology.binary_opening(mask, morphology.disk(apertura)),morphology.disk(chiusura))
    grown_mask = morphology.binary_dilation(closed_mask, morphology.disk(dilatazione))
    return grown_mask

def step_3_highlights_inpainting(img, grown_mask, mask, use_opencv_inpaint: bool = True):
    if use_opencv_inpaint:
        # Metodo 1: cv2.inpaint (Telea) - automatico (performance migliori)
        img_uint8 = (img * 255).astype(np.uint8)
        mask_for_inpaint = (grown_mask * 255).astype(np.uint8)
        inpaint_radius = 5
        restored_image_uint8 = cv2.inpaint(img_uint8, mask_for_inpaint, inpaint_radius, cv2.INPAINT_TELEA)
        img_no_highlights = restored_image_uint8.astype(np.float32) / 255.0
    else:
         # Metodo 2: Riempimento per media
    ''' Identifica le regioni: Il codice trova i pixel che compongono i riflessi e li raggruppa.
    Calcola la media: Per ogni gruppo di pixel, calcola il colore medio dei pixel vicini che non sono né troppo scuri, né troppo chiari (per non falsare la media).
    Riempie l'area: Infine, sostituisce tutti i pixel del riflesso con il colore medio calcolato. '''
        img_hsv = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2HSV)
        _, _, V_channel = cv2.split(img_hsv)
        V_channel = V_channel.astype(np.float32) / 255.0
        # Parametri per escludere i pixel scuri/molto luminosi dalla media
        V_exclude_low = 0.05
        V_exclude_high = 0.95
        img_no_highlights = img.copy()
        label_image_grown = measure.label(grown_mask)
        regions_grown = measure.regionprops(label_image_grown)
        for region_grown_obj in regions_grown:
            valid_pixels_for_mean = []
            for r, c in region_grown_obj.coords:
                current_pixel_color_rgb = img[r, c]
                current_pixel_V = V_channel[r, c]
                is_part_of_original_specular = original_mask[r, c]
                is_valid_intensity = (current_pixel_V > V_exclude_low) and (current_pixel_V < V_exclude_high)
                if not is_part_of_original_specular and is_valid_intensity:
                    valid_pixels_for_mean.append(current_pixel_color_rgb)
            if len(valid_pixels_for_mean) > 0:
                mean_color = np.mean(valid_pixels_for_mean, axis=0)
                for r_orig, c_orig in region_grown_obj.coords:
                    img_no_highlights[r_orig, c_orig] = mean_color
    return img_no_highlights

''' FILTRAGGIO '''

def apply_hybrid_filter(image, bilateral_d=5, bilateral_sigma_color=25, bilateral_sigma_space=50, gaussian_ksize=5, gaussian_sigma=1):
    # Filtro bilaterale
    bilateral_filtered = cv2.bilateralFilter(image, bilateral_d, bilateral_sigma_color, bilateral_sigma_space)
    # Filtro gaussiano
    gaussian_filtered = cv2.GaussianBlur(bilateral_filtered, (gaussian_ksize, gaussian_ksize), gaussian_sigma)
    return gaussian_filtered

def preprocessing_hybrid_filter(image, target_size=(256, 256), crop_percent=0.8, use_clahe=True, apply_median=True):
    img_rgb = image
    normalized = img_rgb.astype(np.float64) / 255.0 # 32 a 64
    # Filtraggio ibrido bilaterale + gaussiano
    #hybrid_filtered = apply_hybrid_filter((normalized * 255).astype(np.uint8))
    hybrid_filtered = apply_hybrid_filter(img_rgb) # si passa di nuovo a uint8, direttamente conviene utilizzare img_rgb comment:02/06
    # CLAHE solo sul canale V (in HSV)
    if use_clahe:
        img_hsv = cv2.cvtColor(hybrid_filtered, cv2.COLOR_RGB2HSV)
        h, s, v = cv2.split(img_hsv)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        v_clahe = clahe.apply(v)
        img_hsv_clahe = cv2.merge([h, s, v_clahe])
        img_rgb_clahe = cv2.cvtColor(img_hsv_clahe, cv2.COLOR_HSV2RGB)
        processed = img_rgb_clahe.astype(np.float64) / 255.0 # 32 a 64 comment:02/06
    else:
        processed = normalized
    return processed  # shape: (H, W, 3), valori in [0,1]

```
